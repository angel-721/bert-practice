```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import pickle
```

```python
from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer
from sklearn.utils import Bunch
from sklearn.datasets import load_files
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import ComplementNB
from  sklearn.model_selection import GridSearchCV, train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = pickle.load(open("../scripts/model.pkl",'rb'))
FEATURES_T = "../train-data/test_features.pkl"
LABELS_T = "../train-data/test_labels.pkl"
CATEGORIES = ['country', 'hip-hop', 'metal', 'pop', 'rock']
test_features = pickle.load(open(FEATURES_T,'rb'))
test_labels = pickle.load(open(LABELS_T,'rb'))
```

```python
model_picks = model.predict(test_features)
cm = confusion_matrix(test_labels, model_picks)
```

```python
display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=CATEGORIES)
display.plot()
plt.show()
```
